```python
from neuromodes.mesh import make_vol_mesh
from nibabel.loadsave import load
import numpy as np
from nilearn import plotting
from lapy import plot

neuromodes = 'C:/Users/ipop0003/neuromodes/neuromodes'
```

```python
# check tetrahedral mesh dimensions
tetmesh = load(f'{neuromodes}/data/hcp_tha-lh_thr25.nii.gz')

dims = tetmesh.shape
nifti_vol = np.prod(dims)
mask_vol = tetmesh.get_fdata().sum()
print(f'NIFTI dimensions: {dims}, volume: {nifti_vol}')
print(f'Volume of mask: {mask_vol} ({mask_vol / nifti_vol * 100:.2f}%)')
```

```python
# plot the nifti file
plotting.plot_img(f'{neuromodes}/data/hcp_tha-lh_thr25.nii.gz', title='Thalamic Mask')
```

```python
tetmesh = make_vol_mesh(f"{neuromodes}/data/hcp_tha-lh_thr25.nii.gz")
print(f'Tetrahedral mesh has {tetmesh.v.shape[0]} vertices and {tetmesh.t.shape[0]} tetrahedra.')
```
